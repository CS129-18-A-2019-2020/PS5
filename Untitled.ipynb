```python
%pylab inline
import pandas as pd
from sklearn import preprocessing, cluster, decomposition
from plotly import express as px
```

```python
df = pd.read_csv('demographic.csv')
```

```python
df.dropna(axis=1).corr()
```

## PCA for Feature Reduction

```python
pca = decomposition.PCA(n_components=10, random_state=42)
```

```python
df_reduced = pd.DataFrame(
    pca.fit_transform(df.dropna(axis=1))
)
```

```python
df_reduced.head()
```

```python
px.bar(
    pd.DataFrame(pca.explained_variance_ratio_, columns=['variance_ratio']),
    y='variance_ratio')
```
